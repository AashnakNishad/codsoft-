```python
a=pd.read_csv("/content/Churn_Modelling.csv")
```

```python
a=a.drop(columns=["RowNumber","CustomerId","Surname"])
```

```python
columns_to_be_encoded=["Geography","Gender"]
```

```python
from sklearn.preprocessing import OneHotEncoder
X = a[columns_to_be_encoded]
encoder = OneHotEncoder()
X_encoded = encoder.fit_transform(X)
X_encoded = pd.DataFrame(X_encoded.toarray(), columns=encoder.get_feature_names_out(columns_to_be_encoded))
a_with_encoded = pd.concat([a, X_encoded], axis=1)
```

```python
a_with_encoded=a_with_encoded.drop(columns=["Geography","Gender"])
```

```python
a_with_encoded
```

```python
X=a_with_encoded.drop(columns=["Exited"])
y=a_with_encoded["Exited"]
```

```python
from sklearn.linear_model import LogisticRegression
my_model=LogisticRegression(random_state=42)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
my_model.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
y_predict=my_model.predict(X_test)
accuracy = accuracy_score(y_test, y_predict)
accuracy
```

```python
from xgboost import XGBClassifier
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
my_model_2=XGBClassifier()
my_model_2.fit(X_train,y_train)
y_predict=my_model_2.predict(X_test)
accuracy = accuracy_score(y_test, y_predict)
accuracy
```
